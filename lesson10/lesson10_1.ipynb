```python
# 請下載日射量json資料,存檔名稱為'台灣日射量.json'
import requests
from requests.exceptions import HTTPError,ConnectionError,Timeout

try:
    url = "https://opendata.cwa.gov.tw/fileapi/v1/opendataapi/O-A0091-001?Authorization=rdec-key-123-45678-011121314&format=JSON"
    response = requests.get(url)
    response.raise_for_status()
    content = response.text
    #print(content)
except ConnectionError:
    print("連線伺服器發生錯誤")
except Timeout:
    print("伺服器忙碌,沒有回應")
except HTTPError:
    print("伺服器回應連線錯誤")
except Exception:
    print("不知名的錯誤")
else:
    with open("台灣日射量.json",encoding="utf-8",mode="w",newline="") as file:
        file.write(content)
    print("存檔完成")
```

```python
import csv
from pprint import pprint
with open('abc.csv',encoding='utf-8',newline='') as file:
    #content = file.read()
    dictReader = csv.DictReader(file)
    contents:list = list(dictReader)

pprint(contents) #List of Dictionarys
```

```python
for item in contents:
    if '元大' in item['證券名稱']:
        print(item)
```
